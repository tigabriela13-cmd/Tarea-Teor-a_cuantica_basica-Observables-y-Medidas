```python
import numpy as np
import math
```

```python
# Definimos hbar (como 1 para simplicidad)
hbar = 1

# Operador de espn
Sx = (hbar/2) * np.array([[0, 1],
                          [1, 0]], dtype=complex)

# Estado inicial |+>_z
spin_up_z = np.array([[1],
                      [0]], dtype=complex)

# Aplicar Sx
psi = Sx @ spin_up_z

# Calcular la probabilidad de medir spin up en la base z
prob_up = np.abs(spin_up_z.conj().T @ psi)**2

print("Estado resultante psi:")
print(psi)
print("\nProbabilidad de medir spin up:")
print(prob_up[0,0].real)
```

```python
import matplotlib.pyplot as plt
# --- parmetros del problema ---
# autovalores (ejemplo tpico)
lambda_vals = np.array([1.0, -1.0])

# estado psi en la base de autovectores (a,b) = (1/sqrt(2), 1/sqrt(2))
a = 1/np.sqrt(2)
b = 1/np.sqrt(2)
probs = np.array([np.abs(a)**2, np.abs(b)**2])  # [1/2, 1/2]

# esperanza y varianza
expectation = np.sum(probs * lambda_vals)
variance = np.sum(probs * (lambda_vals - expectation)**2)

print("Probabilidades p:", probs)
print("Esperanza E[lambda]:", expectation)
print("Varianza Var(lambda):", variance)

# grfico de la distribucin
plt.bar(['lambda1','lambda2'], probs)
plt.ylim(0,1)
plt.ylabel('Probabilidad')
plt.title('Distribucin de probabilidad de los autovalores')
for i,p in enumerate(probs):
    plt.text(i, p+0.02, f"{p:.2f}", ha='center')
plt.show()
```
